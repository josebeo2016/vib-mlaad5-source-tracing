### preprocess

First, you need to download mlaad v5 dataset following this: https://deepfake-total.com/sourcetracing

All audio samples of mlaad v5 and its meta.csv files are located in `./fake/` folder after decompression

```python
# List metadata files in a directory
def find_file(basedir, filename):
    for dirname, dirs, files in os.walk(basedir):
        if filename in files:
            yield os.path.join(dirname, filename)

path=[]
for found in find_file('./fake', 'meta.csv'):
    path.append(found)
print(len(path))
print(type(path[0]))
print(path[0])
```

```python
# Extract metadata and merge into a single DataFrame
dfs = []
for file in path:
    print(file)
    df = pd.read_csv(file,sep='|',quoting=csv.QUOTE_NONE)
    dfs.append(df)

merged_df = pd.concat(dfs)
# fix path value, replace "./" with ""
merged_df['path'] = merged_df['path'].str.replace("\./","")
merged_df.to_csv('meta.csv', index=False)
```

### Devide the data as train, dev, eval

The subset dividing following the original protocol provided by https://deepfake-total.com/data/mlaad4sourcetracing.zip

After decompressing the protocol file of mlaad for source tracing, we got `train.csv, dev.csv, and eval.csv`

```python
# Get the subset information from the protocol
df_train = pd.read_csv('train.csv')
df_train['subset'] = 'train'
df_dev = pd.read_csv('dev.csv')
df_dev['subset'] = 'dev'
df_eval = pd.read_csv('eval.csv')
df_eval['subset'] = 'eval'

merged_df = pd.concat([df_train, df_dev, df_eval])
# merge with df
df = pd.merge(df, merged_df, on='path', how='left')
df
```

```python
df['subset'].value_counts()
```

```python
# remove rows with subset = NaN
df = df.dropna(subset=['subset'])
df
```

```python
# remove model_name_y column, change model_name_x to model_name
df.drop(columns=['model_name_y'], inplace=True)
df.rename(columns={'model_name_x':'model_name'}, inplace=True)
df
```

### Add common voice as real class

CommonVoice dataset can be found here: https://paperswithcode.com/dataset/common-voice

Note that, at the time we wrote this paper, we cannot access to the AILABS for bona-fide samples. 

```python
df_common = pd.read_csv('CommonVoice/meta.csv')
df_common
```

```python
df_common['subset'].value_counts()
```

```python
# select 20% of speakers for train and dev
speakers = df_common['Speaker name'].unique()
np.random.seed(1234)
np.random.shuffle(speakers)
n_train = int(0.2 * len(speakers))
n_dev = int(0.2 * len(speakers))
train_speakers = speakers[:n_train]
dev_speakers = speakers[n_train:n_train+n_dev]
eval_speakers = speakers[n_train+n_dev:]
```

```python
# set the subset for each speaker
df_common['subset'] = 'eval'
df_common.loc[df_common['Speaker name'].isin(train_speakers), 'subset'] = 'train'
df_common.loc[df_common['Speaker name'].isin(dev_speakers), 'subset'] = 'dev'
```

```python
df_common['subset'].value_counts()
```

```python
df_common.to_csv('CommonVoice/meta_ST.csv', index=False)
```

```python
df_common['Language'].value_counts()
```

```python
# combine the two datasets
df = pd.concat([df, df_common])
df['subset'].value_counts()
```

```python
# drop original_file, is_original_language, transcript, utt, Utterence name (file name), TTS or VC, Is multi-speaker?, Noise type 1, Source link, group, Model
df.drop(columns=['original_file', 'is_original_language', 'transcript', 'utt', 'Utterence name (file name)', 'TTS or VC', 'Is multi-speaker?', 'Noise type 1', 'Source link', 'group', 'Model'], inplace=True)
df
```

```python
# calculate the duration of each audio file if it is NaN in duration column
from tqdm import tqdm
import librosa
for i, row in tqdm(full_df.iterrows(), total=len(full_df)):
    if pd.isna(row['duration']):
        try:
            y, sr = librosa.load(row['path'], sr=None)
            duration = librosa.get_duration(y=y, sr=sr)
            full_df.at[i, 'duration'] = duration
        except Exception as e:
            print(f'Error: {e}')
    
```

```python
# make scp folder
import os
from tqdm import tqdm
os.makedirs('scp', exist_ok=True)
# make scp files
full_df = mlaad_df
with open('./scp/bonafide_train.lst', 'w') as f:
    real_train = full_df[(full_df['subset']=='train') & (full_df['label']=='bonafide')]
    for index, row in tqdm(real_train.iterrows(), total=real_train.shape[0]):
        f.write(f'{row["path"]}\n')

with open('./scp/bonafide_dev.lst', 'w') as f:
    real_dev = full_df[(full_df['subset']=='dev') & (full_df['label']=='bonafide')]
    for index, row in tqdm(real_dev.iterrows(), total=real_dev.shape[0]):
        f.write(f'{row["path"]}\n')
        
with open('./scp/spoof_train.lst', 'w') as f:
    spoof_train = full_df[(full_df['subset']=='train') & (full_df['label']=='spoof')]
    for index, row in tqdm(spoof_train.iterrows(), total=spoof_train.shape[0]):
        f.write(f'{row["path"]},{row["acoustic"]},{row["vocoder"]},{row["vocoder2"]},{row["e2e"]}\n')

with open('./scp/spoof_dev.lst', 'w') as f:
    spoof_dev = full_df[(full_df['subset']=='dev') & (full_df['label']=='spoof')]
    for index, row in tqdm(spoof_dev.iterrows(), total=spoof_dev.shape[0]):
        f.write(f'{row["path"]},{row["acoustic"]},{row["vocoder"]},{row["vocoder2"]},{row["e2e"]}\n')

with open('./scp/eval.lst', 'w') as f:
    full_eval = full_df[full_df['subset']=='eval']
    for index, row in tqdm(full_eval.iterrows(), total=full_eval.shape[0]):
        f.write(f'{row["path"]},{row["acoustic"]},{row["vocoder"]},{row["vocoder2"]},{row["e2e"]}\n')
```
